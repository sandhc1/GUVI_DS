```python
import pandas as pd
data1 = {
    'ID' : ['1','2','3','4','5'],
    'Feature1' : ['A','B','C','D','E'],
    'Feature2' : ['L','K','N','O','P']
}

df1 = pd.DataFrame(data1, columns = ['ID','Feature1','Feature2'])
df1.head()
```

```python
data2 = {
    'ID' : ['10','21','32','43','56'],
    'Feature1' : ['AZ','BC','CV','DB','EW'],
    'Feature2' : ['LJ','KH','NF','OG','PQ']
}

df2 = pd.DataFrame(data2, columns = ['ID','Feature1','Feature2'])
df2.head()
```

```python
data3 = {
    'ID11' : ['109','21f','325','434','563'],
    'Feature11' : ['AZ','BCsd','CVsd','DdfB','EWfg'],
    'Feature22' : ['LJs','KHf','NFf','OGr','PQty']
}

df3 = pd.DataFrame(data3, columns = ['ID11','Feature11','Feature22'])
df3.head()
```

```python
#CONCAT

dfa = pd.concat([df1,df2],ignore_index=True)
dfa
```

```python
dfb = pd.concat([df1,df2],axis=1)
dfb
```

```python
#join

df1.join(df3)
```

```python
data1 = pd.read_csv("https://raw.githubusercontent.com/nethajinirmal13/Training-datasets/main/dw21%20-%20Sheet1.csv")
data2 = pd.read_csv("https://raw.githubusercontent.com/nethajinirmal13/Training-datasets/main/dw21-set2%20-%20Sheet1.csv")
```

```python
data1.head()
```

```python
data2.head()
```

```python
data1.shape

```

```python
data2.shape
```

```python
#outer merge
data_o = pd.merge(data1,data2,on="State/UT",how = 'outer')
data_o
```

```python
# inner merge
data_i = pd.merge(data1,data2,on="State/UT",how = 'inner')
print(data_i.head())
```

```python
# right merge
data_r = pd.merge(data1,data2,on="State/UT",how = 'right')
data_r
```

```python
# left merge
data_l = pd.merge(data1,data2,on="State/UT",how = 'left')
data_l
```

```python
#join

data1.join(data2['Population']).join(data2['Increase'])
```

```python
#concat

pd.concat([data1,data2],ignore_index=True).sample(10)
```

```python
#titanic data

import pandas as pd
data = pd.read_csv("https://raw.githubusercontent.com/nethajinirmal13/Training-datasets/main/titanic_train.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
data.groupby('Sex').ngroups
```

```python
data.groupby('Sex').size()
```

```python
data.groupby('Survived').PassengerId.agg(['count'])
```

```python
data.groupby('Pclass').agg(['count','max','min'])
```

```python
data.head()
```

```python
data.groupby(['Survived','Pclass','Sex']).PassengerId.agg(['count'])
```

```python
#pivot table

pd.pivot_table(data, index = ['Sex','Pclass'], aggfunc='count')
```
